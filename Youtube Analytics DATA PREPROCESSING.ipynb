```python
import pandas as pd
import json
```

```python
file = 'tanmay_bhat.json'
data = None
with open(file, 'r') as f:
    data = json.load(f)

channel_id, stats = data.popitem()
```

```python
print(channel_id)
```

```python
channel_stats = stats["channel_statistics"]
video_stats = stats["video_data"]
```

```python
#channel statistics

print('Total Views', channel_stats["viewCount"])
print('Subscribers', channel_stats["subscriberCount"])
print('Total Videos', channel_stats["videoCount"])
```
