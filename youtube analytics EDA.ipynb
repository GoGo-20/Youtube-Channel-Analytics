```python
import pandas as pd
import matplotlib as 
```

```python
df = pd.read_csv('youtubeAnalyticsData.csv')
```

```python
df.head()
```

```python
df = df.sort_values(by = ['viewCount'],ascending=False)
```

```python
top10 = df.head(10)
top10
```

```python
ax = top10.plot.bar(x = 'title', y= 'viewCount', figsize=(12,8), fontsize = 14)
```

```python
bottom10 = df.tail(10)
bottom10
```

```python
ax = bottom10.plot.bar(x = 'title', y= 'viewCount', figsize=(12,8), fontsize = 14)
```

```python
#likes/comments
likes_avg = top10['likeCount'].mean()
comment_avg = top10['commentCount'].mean()
print('Top 10')
print('average likes', likes_avg)
print('average comments', comment_avg)
```
